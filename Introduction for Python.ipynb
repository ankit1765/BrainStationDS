```python
import numpy as np 
import matplotlib.pyplot as plt

#Steps = Maxmimu number on scale 
#Repeats Counting by 10 
steps, repeats = 100, 10 
stepstaken = 0.01 * np.random.randn(steps,repeats)
positions = stepstaken.cumsum (axis = 0)
plt.plot(positions);
plt.xlabel('time');
plt.ylabel('position');
```
